*Para una capa convolucional, el número de parámetros se calcula como:*

# ((ancho del filtro × alto del filtro × canales de entrada)pesos por filtro) x canales de salida(num filtros) + canales de salida(sesgos)

No depende del tamaño de la entrada 128×128, sino de cuántos canales entran y cuántos mapas de características salen.

# Capa 1

**Tamaño de filtro:** 3×3  
**Canales de entrada:** 3 (RGB)  
**Canales de salida:** 100  

**Parámetros:**  
\((3×3×3)×100+100 = (9×3)×100+100 = 27×100+100 = 2700+100 = 2800\)

---

# Procedimiento de una capa convucional:

- Primero la recepción de entrada, la capa obtiene una imagen o mapa de la capa anterior en formato (altura, ancho, canales)

- Se aplican los filtros, se definen kernels, los de 3 x 3 o 5 x 5, cada filtro sobre la imagen multiplicando y sumando valores con sus pesos

- Se añade el sesgo, un valor que ajusta la salida del filtro antes de la activación

- Se aplica la función de activación, como ReLU, eliminando valores negativos y mejorando el aprendizaje

- Se genera un nuevo mapa de características, resaltando patrones como bordes o texturas en la imagen


# Utilidad

- Básicamente extrae patrones de una imagen de manera eficiente, reduciendo parámetros, y así es capaz de reconocer objetos sin que importe la posición
